```python
# Simple ReAct Agent (VERY EASY)

def calculator(exp):
    return str(eval(exp))

def agent(query):
    print("User:", query)

    # THINK
    if any(x in query for x in ["+", "-", "*", "/"]):
        print("Thought: This is a math question")

        # ACT
        expression = query.replace("What is", "").strip().replace("?", "")
        print("Action: Using calculator")
        print("Input:", expression)

        result = calculator(expression)

        # OBSERVE
        print("Observation:", result)

        # FINAL ANSWER
        print("Final Answer:", result)

    else:
        print("Thought: This is general question")
        print("Action: Using search")
        print("Observation: Example search result")
        print("Final Answer: Example result")

# Run it
agent("What is 6 * 10?")
```
